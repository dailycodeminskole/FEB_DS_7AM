```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# BOX PLOT

```python
iris = sns.load_dataset('iris')
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
plt.show()
```

# VIOLEN PLOT

```python
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.show()
```
