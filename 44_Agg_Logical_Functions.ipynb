# Logical Functions

```python
# any : True even if one elemnt is True
# all : True only  if all elemnt are True
```

```python
arr4 = np.array([1,0,0,0,0])
```

```python
arr4
```

```python
np.any(arr4)
```

```python
arr5 = np.array([1,1,1,1,0])
```

```python
np.all(arr5)
```

```python
arr5 = np.array([1,1,1,1,1])
np.all(arr5)
```
