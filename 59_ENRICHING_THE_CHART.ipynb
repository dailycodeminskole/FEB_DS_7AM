```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# plotting square functions
x_val = np.arange(-5,5,.1)
# y_val = [25,16,...0,1,4,9,16,25]
y_val = x_val**2
# plt.xticks(np.arange(-5,5,.5))
plt.xticks(np.arange(-5,5,.5),rotation=90)

plt.plot(x_val,y_val,color='r') #changing the default color
plt.show()
```

```python
# plotting square functions
x_val = np.arange(-5,5,.1)
# y_val = [25,16,...0,1,4,9,16,25]
y_val = x_val**2
# plt.xticks(np.arange(-5,5,.5))
plt.xticks(np.arange(-5,5,.5),rotation=90)

plt.plot(x_val,y_val,'r*') #red star
plt.show()
```

# Plotting multiple plots in the same canvas

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
plt.show()
```

# Adding Legends

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val,label="Square")
plt.plot(x_val,y2_val,label="Cube")
plt.legend()
plt.show()
```

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
plt.legend(["Square","Cube"],loc='lower right',frameon=False,ncol=2)
plt.show()
```

```python
x_val = np.arange(-10,10,.5)
y1_val = x_val**2
y2_val = x_val**3
plt.plot(x_val,y1_val)
plt.plot(x_val,y2_val)
plt.legend(["Square","Cube"],loc='lower right',frameon=False,ncol=2)

plt.xlim(-2.5,2.5)
plt.ylim(-10,10)
plt.show()
```
