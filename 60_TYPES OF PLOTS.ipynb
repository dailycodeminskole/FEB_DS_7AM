```python
import numpy as np
import matplotlib.pyplot as plt
```

# BAR

```python
# shop1
# fruit 1 : 10
# fruit 2 : 20
# fruit 3 : 5
```

```python
plt.bar([1,2,3],[10,20,5])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=.3)
plt.show()
```

```python
# shop2
# fruit 1 : 5
# fruit 2 : 10
# fruit 3 : 6
```

```python
plt.bar([1,2,3],[5,10,6])
plt.show()
```

```python
# plotting both shops data in one bar chart
```

```python
plt.bar([1,2,3],[10,20,5],width=.5)
plt.bar([1,2,3],[5,10,6],width=.2)
plt.legend(['Shop1','Shop2'])
plt.show()
```

```python
plt.bar([1,2,3],[10,20,5],width=.2)
plt.bar([1.2,2.2,3.2],[5,10,6],width=.2)# shifting the x axis to plot the 2nd shop data
plt.legend(['Shop1','Shop2'])
plt.show()
```
